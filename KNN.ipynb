```python
import pandas as pds
import numpy as ny
import csv
```

```python
def loadDataTrain():
    filename = 'DataTrain.csv'
    datatr = []
    with open(filename) as f:
        file = csv.reader(f)
        next(file)
        for row in file:
            datatr.append([int(row[0]),float(row[1]),float(row[2]),float(row[3]),float(row[4]),float(row[5]),int(row[6])])
    return datatr
```

```python
def loadDataTest():
    filename = 'DataTest.csv'
    datats = []
    with open(filename) as f:
        file = csv.reader(f)
        next(file)
        for row in file:
            datats.append([int(row[0]),float(row[1]),float(row[2]),float(row[3]),float(row[4]),float(row[5])])
    return datats
```

```python
def dist(a,b):
    s = 0
    for i in range (len(a)):
        s += ((a[i]-b[i])**2)
    return (ny.sqrt(s))
```

```python
def vote(dtrain, nearest_neighbor):
    votes = [0,0,0,0]
    for i in range (len(nearest_neighbor)):
        classNN = dtrain[nearest_neighbor[i][1]][-1] #get class from nearest_neighbor[i]
        votes[classNN] += 1
    class_test = votes.index(max(votes))
    return (class_test)
    
```

```python
if __name__ == '__main__':
    datats = loadDataTest()
    datatr = loadDataTrain()
    K = 3
    class_of_data_test = []
    for i in range (len(datats)):
        distances=[]
        for j in range (len(datatr)):
            jarak=dist(datats[i][1:6],datatr[j][1:6])
            distances.append([jarak,j])
        solution =  sorted(distances,reverse=False)
        class_of_data_test.append(vote(dtrain=datatr,nearest_neighbor=solution[:K]))

    for i,value in enumerate(datats):
        print ('{}, class: {}'.format(datats[i],class_of_data_test[i]))
```
